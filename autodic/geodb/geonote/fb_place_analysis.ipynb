```python
import numpy as np
from numpy import NAN as na

import json
import pandas as pd
from pandas import DataFrame, Series

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = json.load(open('../fb/place.json'))
```

```python
len(data['_default'])
```

```python
data['_default']['11542']
```

```python
records = [v for v in data['_default'].itervalues()]
```

```python
df = DataFrame(records, columns=['id', 'name'])
```

```python
df
```

```python
import re
import mojimoji
re_parenthes = re.compile(ur'（.+?）')
df['norm'] = df['name'].apply(lambda x: mojimoji.han_to_zen(re_parenthes.sub('', x.upper())))
#df['norm'] = df['name'].apply(lambda x: mojimoji.han_to_zen(x))
```

```python
df['a'] = df['norm'].apply(lambda x: x.split()[0] if 0 < len(x.split()) else '')
df['b'] = df['norm'].apply(lambda x: x.split()[1] if 1 < len(x.split()) else '')
df['c'] = df['norm'].apply(lambda x: x.split()[2] if 2 < len(x.split()) else '')
df['d'] = df['norm'].apply(lambda x: x.split()[3] if 3 < len(x.split()) else '')
```

```python
df
```

```python
df.norm
```

```python
a = df.a.value_counts()
b = df.b.value_counts()
c = df.c.value_counts()
d = df.d.value_counts()

new_index = a.index | b.index | c.index | d.index

a = a.reindex(new_index, fill_value=0)
b = b.reindex(new_index, fill_value=0)
c = c.reindex(new_index, fill_value=0)
d = d.reindex(new_index, fill_value=0)

total_count = a + b + c + d
total_count.sort_values(ascending=False)[:1000]
```

```python
a = df.a[df.a.str.endswith(u'店')]
b = df.b[df.b.str.endswith(u'店')]
c = df.c[df.c.str.endswith(u'店')]
d = df.d[df.d.str.endswith(u'店')]

a = a.value_counts()
b = b.value_counts()
c = c.value_counts()
d = d.value_counts()

new_index = a.index | b.index | c.index | d.index

a = a.reindex(new_index, fill_value=0)
b = b.reindex(new_index, fill_value=0)
c = c.reindex(new_index, fill_value=0)
d = d.reindex(new_index, fill_value=0)

total_count = a + b + c + d

total_count.sort_values(ascending=False)
```
