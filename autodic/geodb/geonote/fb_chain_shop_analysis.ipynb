```python
import numpy as np
from numpy import NAN as na

import json
import pandas as pd
from pandas import DataFrame, Series
```

```python
data = json.load(open('../fb/place.json'))
```

```python
data = [v for v in data['_default'].itervalues()]
```

```python
import re
import mojimoji

from collections import defaultdict
```

```python
# stats category
category_dict = defaultdict(int)
for row in data:
    category_dict[row['category_list'][0]['name']] += 1
#    if row['category_list'][0]['name'] == 'Park':
#        print row['name']
        
for k, v in sorted(category_dict.items(), key=lambda x: x[1], reverse=True):
    print k, v
```

```python
# ○○ ●●店
# ○○
shop_name_counts = defaultdict(int)
shop_name_category = defaultdict(lambda: defaultdict(int))
# ●●店
chain_suffix_counts = defaultdict(int)

re_paren = re.compile(ur'（.+?）')
re_space = re.compile(ur'[　]+')
for row in data:
    name = mojimoji.han_to_zen(row['name'])
    name = re_paren.sub('', name)
    name = re_space.sub(u'　', name)
    name = name.strip(u'　')
    last_sepalation = max(name.rfind(u'　'), name.rfind(u'／'))
    if not last_sepalation == -1 and name.endswith(u'店'):
        shop_name_counts[name[:last_sepalation]] += 1
        chain_suffix_counts[name[last_sepalation+1:]] += 1
        shop_name_category[name[:last_sepalation]][row['category_list'][0]['name']] += 1

shop_names = shop_name_counts.keys()
chain_suffixes = chain_suffix_counts.keys()
```

```python
for name in shop_names:
    print name
```

```python
for suffix in chain_suffixes:
    print suffix
```

```python
# Explore more using existing shop ans chain list
# ○○●●店
# ○○
new_shop_name_counts = defaultdict(int)
# ●●店
new_chain_suffix_counts = defaultdict(int)

re_paren = re.compile(ur'（.+?）')
for row in data:
    name = mojimoji.han_to_zen(row['name'])
    name = re_paren.sub('', name)
    name = re_space.sub(u'　', name)
    name = name.strip(u'　')
    last_sepalation = max(name.rfind(u'　'), name.rfind(u'／'))
    
    if last_sepalation == -1 and name.endswith(u'店'):
        for shop_name in shop_names:
            if len(shop_name) < len(name) and name.startswith(shop_name):
                if not name[len(shop_name):] in chain_suffix_counts:
                    chain_suffix_counts[name[len(shop_name):]] += 1
                    new_chain_suffix_counts[name[len(shop_name):]] += 1
                    
        for suffix in chain_suffixes:
            if len(suffix) < len(name) and name.endswith(suffix):
                if not name[:-len(suffix)] in shop_name_counts:
                    shop_name_counts[name[:-len(suffix)]] += 1
                    new_shop_name_counts[name[:-len(suffix)]] += 1
                    shop_name_category[name[:-len(suffix)]][row['category_list'][0]['name']] += 1
```

```python
for name in new_shop_name_counts.keys():
    print name
```

```python
for chain in new_chain_suffix_counts.keys():
    print chain
```

```python
def best_category(name):
    if name in shop_name_category and 0 < len(shop_name_category[name]):
        return sorted(shop_name_category[name].items(), key=lambda x: x[1], reverse=True)[0][0]
    else:
        None
```
